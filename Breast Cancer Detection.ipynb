**Train and Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
#test size-->to specify th eprecentage of test data needed
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y)
#stratify --> for correct distributioon of data as of th orignal
```

```python
print(y.mean(),y_train.mean(),y_test.mean())
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
#random_state-->specific split of data each value of random_state splits the data differently
```

```python
print(x_train.mean(),x_test.mean(),x.mean())
```

```python
print(x_train)
```

Logistic Regression Binary Classification

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier =LogisticRegression()
#loading the logistic regression model to the variable "classifier"
```

```python
#training the model on training data
classifier.fit(x_train,y_train)
```

Evaluation of the Model

```python
#import accuracy_score
from sklearn.metrics import accuracy_score
```

```python
prediction_on_training_data = classifier.predict(x_train)
accuracy_on_training_data=accuracy_score(y_train,prediction_on_training_data )
```

```python
print("Accuracy on training data:",accuracy_on_training_data)
```

```python
#prediction on test_data
prediction_on_test_data=classifier.predict(x_test)
accuracy_on_test_data=accuracy_score(y_test,prediction_on_test_data )
```

```python
print("Accuracy on test data:",accuracy_on_test_data)
```

Detecting Whether the patient has breast camncer in benign or Malignant stage

```python
#change the input_data to numpy_ar to make prediction
input_data=(9.504,12.44,60.34,273.9,0.1024,0.06492,0.02956,0.02076,0.1815,0.06905,0.2773,0.9768,1.909,15.7,0.009606,0.01432,0.01985,0.01421,0.02027,0.002968,10.23,15.66,65.13,314.9,0.1324,0.1148,0.08867,0.06227,0.245,0.07773)
input_data_as_numpy_array=np.asarray(input_data)
print(input_data)
#reshape the array as we are predicting the ouput for one instanec
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
#prediction
prediction=classifier.predict(input_data_reshaped)
print(prediction) #returns alist with element 0-Malignant, 1-Benign

if prediction==0:
    print("The breast Cancer is Malignant")
else:
    print("The breast Cancer is Benign")
```
